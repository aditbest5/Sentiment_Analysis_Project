Topik yang diangkat adalah analisis sentimen terhadap produk indomie. Seperti yang diketahui kebanyakan orang, Indomie merupakan produk yang sangat populer di kalangan masyarakat pecinta mie. Kita akan mencari tahu sentimen masyarakat tentang indomie apakah masyarakat di twitter memiliki sentimen yang positif terhadap produk satu ini atau malah mereka berpandangan negatif terhadap produk indomie ini.

```python
# import semua library dan package

import pandas as pd, numpy as np, matplotlib.pyplot as plt
import json, tweepy, requests, re, nltk, string
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords

from bs4 import BeautifulSoup as bs
from urllib.request import urlopen, Request
import pandas as pd
```

```python
#membuka kode token API kita
with open("token.json")as f:
  tokens = json.load(f)

bearer_token = tokens['bearer_token']
api_key = tokens['api_key']
api_key_secret = tokens['api_key_secret']
access_token = tokens['access_token']
access_token_secret = tokens['access_token_secret']
```

```python
#Kata kunci indomie dipilih karena kita ingin mengambil tweet netizen twitter tentang indomie
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

search_word = "indomie"
date_since = "2022-01-23"
date_until = "2022-01-26"
new_search = search_word + "-filter:retweets"

tweets = tweepy.Cursor(api.search,
                       q=new_search,
                       lang="id",
                       since=date_since,
                       until=date_until).items(1000)
```

```python
#Menampilkan hasil tweet
items = []
for tweet in tweets:
  items.append(tweet.text)

items
```

```python
#Mengubah hasil tweet ke dalam dataframe dengan kolom title
hasil = pd.DataFrame(data=items, columns=['title'])
hasil
```

```python
#Export hasil data ke csv
hasil.to_csv('indomie_tweet_search.csv',index=False)
```
